```python
import pandas as pd
from sqlalchemy import create_engine
import urllib.parse as up
import json
from pandas.io.json import json_normalize

# Read the cleaned data from Part 1
title_basics = pd.read_csv('title_basics-Copy1.csv')
title_ratings = pd.read_csv('title_ratings.csv')

# Read the TMDB API data
with open('Data/tmdb_api_results_2000.json', 'r') as file:
    tmdb_data_2000 = json.load(file)

with open('Data/tmdb_api_results_2001.json', 'r') as file:
    tmdb_data_2001 = json.load(file)

# Concatenate the TMDB API data for both years
tmdb_data = tmdb_data_2000 + tmdb_data_2001

# Flatten the dictionaries within tmdb_data
flattened_tmdb_data = []
for item in tmdb_data:
    if isinstance(item, dict):
        flattened_tmdb_data.append(json_normalize(item))

# Concatenate the flattened data into a single DataFrame
if flattened_tmdb_data:
    tmdb_data_df = pd.concat(flattened_tmdb_data, ignore_index=True)
else:
    tmdb_data_df = pd.DataFrame()

# Replace the placeholders with your MySQL connection details
username = 'root'
password = up.quote_plus('OscarRamirez123@')
host = 'localhost'
database = "movies"

# Create the connection string
connection_str = f'mysql+pymysql://{username}:{password}@{host}/{database}'

# Create the engine
engine = create_engine(connection_str)

# Create the title_basics table
title_basics.to_sql('title_basics', engine, if_exists='replace', index=False)

# Create the title_ratings table
title_ratings.to_sql('title_ratings', engine, if_exists='replace', index=False)

# Create the tmdb_data table
tmdb_data_df.to_sql('tmdb_data', engine, if_exists='replace', index=False)

# Show the first 5 rows of each table
print("title_basics:")
print(pd.read_sql_query('SELECT * FROM title_basics LIMIT 5', engine))
print("title_ratings:")
print(pd.read_sql_query('SELECT * FROM title_ratings LIMIT 5', engine))
print("tmdb_data:")
print(pd.read_sql_query('SELECT * FROM tmdb_data LIMIT 5', engine))

# Show all tables in the database
print("Tables in the database:")
print(pd.read_sql_query('SHOW TABLES', engine))
```
